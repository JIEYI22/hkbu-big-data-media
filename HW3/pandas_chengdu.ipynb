```python
import plotly.plotly as py
import plotly.graph_objs as go
plotly.offline.init_notebook_mode(connected=True)
from plotly.graph_objs import Scatter, Layout
import numpy as np
from plotly.graph_objs import Bar
from plotly.offline import iplot
```

we extracted some kinds of attractions based on keywords provided by Chengdu local people from whose perspective these types can be representative of Chengdu's sceneries.

```python
园林类=df['spot'].apply(lambda t:'园' in str(t)).aggregate('sum')
古镇=df['spot'].apply(lambda t:'古镇' in str(t)).aggregate('sum')
博物馆=df['spot'].apply(lambda t:'博物馆'in str(t)).aggregate('sum')
庭院=df['spot'].apply(lambda t:'院'in str(t)).aggregate('sum')
寺=df['spot'].apply(lambda t:'寺'in str(t)).aggregate('sum')
庙=df['spot'].apply(lambda t:'庙'in str(t)).aggregate('sum')
广场=df['spot'].apply(lambda t:'广场' in str(t)).aggregate('sum')
```

```python
山=df['spot'].apply(lambda t:'山' in str(t)).aggregate('sum')
沟=df['spot'].apply(lambda t:'沟' in str(t)).aggregate('sum')
溪=df['spot'].apply(lambda t:'溪' in str(t)).aggregate('sum')
温泉=df['spot'].apply(lambda t:'温泉' in str(t)).aggregate('sum')
湖=df['spot'].apply(lambda t:'湖' in str(t)).aggregate('sum')
```

```python
data = [go.Bar(
    x=['园林类', '古镇', '博物馆','庭院','寺','庙','广场'],
    y=[园林类,古镇,博物馆,庭院,寺,庙,广场],
    marker=dict(
        color=['rgba(222,45,38,0.8)', 'rgba(204,204,204,1)',
               'rgba(204,204,204,1)', 'rgba(204,204,204,1)',
               'rgba(204,204,204,1)','rgba(204,204,204,1)','rgba(204,204,204,1)']),
)]
iplot({
    "data": data,
    "layout": Layout(title="Human Landscape")
})
```

The most classic way to spend leisure time for Chengdu people is to sit in parks or gardens, chatting with family or friends, enjoying tea-tasting and playing Majiang. So parks and gardens as you can see make up the largest part of human landscape.

The landform of Chengdu is varied for its long and complicated tectonic activity，which also leads to diverse types of natural landscape.
Mountains contribute a large propotion to natural landscape.Being at the central of Sichuan Basin, Chengdu Plain is always a base camp for mountaineering enthusiasts,from where they can get to different kinds of mountains,including Mount Qingcheng,the birthplace of Taoism, the Xiling Snow Mountain that once described by Dufu,and also Minya Konka,which is the easternmost 7,000 metres (23,000 ft) peak in the world and be known as "the King of Sichuan Mountains".

```python
df_top_100=df.sort_values(by='recommend',ascending=False)[:100].reset_index()
```

```python
df_top_100
```

```python
data = [go.Scatter(
    x = df_top_100.index,
    y = df_top_100['rank'].values,
    mode = 'lines+markers',
    name = 'recommend'   
)]
iplot({"data":data, 
       "layout": Layout(title="scores of comments_top_100 spot")
      })
```

we sorted all scenic spots from high to low according to the number of comments,finding out the top 100 most popular spots in Chengdu, within which we continue to analysis the quality of scenic spots by checking scores.  
We find out that not all spots with a large amount of comments get high scores at the same time. 

```python
top_100_recommend=df_top_100[df_top_100['rank']>=4.7]
```

```python
top_100_recommend.sort_values(by='rank',ascending=False)
```

Apart from those well-known spots like Dujiangyan and Xiling Snow Mountain,we introduce some sleepers of Chengdu attractions which may not be very famous but you still can have a goodtime there（scores higher than 4.7), such as the ski resort located in Xindu(4.9) and the Shufenyayun Theatre where you can enjoy Sichuan Opera(4.8 and 1935comments).
