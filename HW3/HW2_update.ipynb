```python
import requests
import bs4
import csv
from bs4 import BeautifulSoup
```

```python
def get_the_page(url):  
    r=requests.get(url)
    mypage=BeautifulSoup(r.text)

    myspot=mypage.find_all('div',attrs={'class':'rdetailbox'})
    locations=[]
    for location in myspot:
        myl=location.find('a')
        locations.append(myl.text)

    myscores=mypage.find_all('ul',attrs={'class':'r_comment'})
    scores=[]
    for score in myscores:
        myscore=score.find('strong')
        scores.append(myscore.text)
    
    myrecommend=mypage.find_all('a',attrs={'class':'recomment'})
    recommends=[]
    for recommend in myrecommend:
        rm=int(recommend.text.strip().lstrip('(').rstrip('条点评)'))
        recommends.append(rm)
        

    gps=mypage.find_all('dd',attrs={'class':'ellipsis'})
    GPS=[]
    for GPSs in gps:
        GPS.append(GPSs.text.strip())
    
    
    
    page_content=[]
    for i in range (0,len(locations)):
        page_content.append([locations[i],recommends[i],scores[i],GPS[i]])
    
    return page_content
```

```python
url_list=[]
for i in range(0,50):
    page_number=str(i+1)
    url = 'http://you.ctrip.com/sight/chengdu104/s0-p' + page_number +'.html'
    url_list.append(url)
```

```python
information=[]
for spot in url_list:
    my=get_the_page(spot)
    information.extend(my)
```

```python
with open('chengdupage_recommend.csv', 'w', newline='') as csvfile:
    spamwriter = csv.writer(csvfile, delimiter=',',
                            quotechar='"', quoting=csv.QUOTE_MINIMAL)
    header= ['spot','recommend','rank','satellite']
    spamwriter.writerow(header)
    spamwriter.writerows(information)
```
